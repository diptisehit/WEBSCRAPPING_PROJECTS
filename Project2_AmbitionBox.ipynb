# PROJECT - SCRAPPING CONTENTS OF 'AMBITION BOX' WEBSITE FOR LIST OF IT COMPANIES

- Webscrapping is used to extract the front-end data of website for analysis.
- Packages used for webscrapping are - BeautifulSoup and requests.
- requests is used to encode the response received based on the HTTP headers.
- BeautifulSoup is used for pulling data out of HTML and XML file and then navigating, searching and modifying the parse tree.


# Importing python libraries for scrapping

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Getting [403] response :

- It means the website is restricted.
- So we will use headers to get response 200.

```python
# headers content to get access
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
html=requests.get(url,headers=headers)
print(url)
print(html)
```

```python
bsoup=BeautifulSoup(html.text,'lxml')
bsoup
```

```python
companyName=[]
for page in range(1,20):
    html = requests.get(url + '?page=' + str(page),headers=headers) 
    bsoup=BeautifulSoup(html.text,'lxml')
    divlist=bsoup.find_all('div',attrs={"class":"companyCardWrapper__metaInformation"})
    for div in divlist:
        for i in div:
            companyName.append(i[0].getText())
    
```

# Defining Function for scrapping multiple pages

```python
def companyDetailsFunc(pages):
    url= f'https://www.ambitionbox.com/it-services-and-consulting-companies-in-india?page={pages}'
    html=requests.get(url,headers=headers)
    bsoup=BeautifulSoup(html.text,'lxml')
    divlist=bsoup.find_all('div',attrs={"class":"companyCardWrapper__metaInformation"})
    
    for company in divlist:
        companyDetails={
        'companyName':company.h2.getText().strip(),
        'companyRating':company.find('span',attrs={"class":"companyCardWrapper__companyRatingValue"}).getText(),
        'ageOfCompany':company.find('span',attrs={"class":"companyCardWrapper__interLinking"}).getText().split('|')[1].replace('Employees','').strip(),
        #'HighlyRatedFor':any(company.find('span',attrs={"class":"companyCardWrapper__ratingValues"}).getText())
        }
        companyList.append(companyDetails)
    return



companyList=[]

for i in range(20):
    companyDetailsFunc(i)
    
companyList
    
```
